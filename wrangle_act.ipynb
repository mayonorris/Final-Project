##### Détection de valeurs abérrantes: méthode
Les valeurs qui sont inférieures à $Q1-1.5*IQR$ ou suppérieures à $Q3+1.5*IQR$ seront considérées comme des valeurs abérrantes. 
+ Q1: premier quartile
+ Q3: troisième quartile 
+ IQR: intervalle interquartile

```python
# Le code suivant permet de détecter les valeurs abérrantes: 
def find_outliers(x):
    q1=x.quantile(0.25)
    q3=x.quantile(0.75)
    iqr=q3-q1
    out_list=[]
    for val in x:
        if (val<q1-1.5*iqr) or (val>q3+1.5*iqr):
            out_list.append(val)
    return (q1,q3,iqr, q1-1.5*iqr, q3+1.5*iqr),pd.DataFrame(out_list, columns=['Outliers'])
```
