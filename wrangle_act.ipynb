##### D√©tection de valeurs ab√©rrantes: m√©thode
Les valeurs qui sont inf√©rieures √† $Q1-1.5*IQR$ ou supp√©rieures √† $Q3+1.5*IQR$ seront consid√©r√©es comme des valeurs ab√©rrantes. 
+ Q1: premier quartile
+ Q3: troisi√®me quartile 
+ IQR: intervalle interquartile

```python
# Le code suivant permet de d√©tecter les valeurs ab√©rrantes: 
def find_outliers(x):
    q1=x.quantile(0.25)
    q3=x.quantile(0.75)
    iqr=q3-q1
    out_list=[]
    for val in x:
        if (val<q1-1.5*iqr) or (val>q3+1.5*iqr):
            out_list.append(val)
    return (q1,q3,iqr, q1-1.5*iqr, q3+1.5*iqr),pd.DataFrame(out_list, columns=['Outliers'])
```
